```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime

df_raw=pd.read_csv("/Users/guywinfield/PycharmProjects/Data/NYC hourly traffic.csv")


df = df_raw.copy()
```

```python
df["Date"]=df["Date"].apply(pd.to_datetime)

df.head()
```

```python
sample=df.query("`Plaza ID`==5 & Date>'2016-12-30' & Date<'2017-05-01'")
sample
```

```python
sample = df[(df["Plaza ID"]==5) & (df["Date"]>'2016-12-30') & (df["Date"]<'2017-05-01')]
sample
```

```python
plt.hist(sample["Hour"],bins=24,weights=sample["# Vehicles - E-ZPass"]);
```

```python
sample["Day of Week"] = sample["Date"].dt.dayofweek

plt.hist(sample["Day of Week"],bins=7,weights=sample["# Vehicles - E-ZPass"]);
```

```python
plt.figure(figsize=(12,8))
plt.hist2d(sample["Hour"],sample["Day of Week"],bins=[24,7], weights=sample["# Vehicles - E-ZPass"])

plt.set_cmap('plasma')
plt.colorbar(),
```
